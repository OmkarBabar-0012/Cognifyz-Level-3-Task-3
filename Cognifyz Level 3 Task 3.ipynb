# Task: Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\91702\Dataset.csv'
df = pd.read_csv(file_path)

# Task 1: Visualizations to represent the distribution of ratings using different charts

# Histogram of Ratings (no need to limit to top 10)
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Bar Plot of Ratings (no need to limit to top 10)
plt.figure(figsize=(10, 6))
sns.countplot(x='Aggregate rating', data=df, palette='viridis')
plt.title('Bar Plot of Ratings', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Task 2: Compare the average ratings of different cuisines or cities (Top 10)

# Top 10 Cuisines by Average Rating
top_10_cuisines = df.groupby('Cuisines')['Aggregate rating'].mean().nlargest(10).index
df_top_cuisines = df[df['Cuisines'].isin(top_10_cuisines)]

plt.figure(figsize=(12, 6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=df_top_cuisines, palette='muted', ci=None)
plt.title('Top 10 Cuisines by Average Rating', fontsize=16)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Cuisine', fontsize=14)
plt.ylabel('Average Rating', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Top 10 Cities by Average Rating
top_10_cities = df.groupby('City')['Aggregate rating'].mean().nlargest(10).index
df_top_cities = df[df['City'].isin(top_10_cities)]

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Aggregate rating', data=df_top_cities, palette='coolwarm', ci=None)
plt.title('Top 10 Cities by Average Rating', fontsize=16)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('City', fontsize=14)
plt.ylabel('Average Rating', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Task 3: Visualize the relationship between various features and the target variable (Top 10)

# Scatter plot between Average Cost for Two and Rating (Top 10 Cities)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df_top_cities, hue='City', palette='deep')
plt.title('Relationship between Average Cost for Two and Rating (Top 10 Cities)', fontsize=16)
plt.xlabel('Average Cost for Two', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Pairplot to visualize relationships between multiple features (Top 10 Cuisines)
sns.pairplot(df_top_cuisines, hue='Cuisines', palette='husl')
plt.suptitle('Pairplot of Features in the Dataset (Top 10 Cuisines)', y=1.02, fontsize=16)
plt.show()

# Heatmap to show correlations between features (Top 10 Cities)
plt.figure(figsize=(12, 8))
sns.heatmap(df_top_cities.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Feature Correlations (Top 10 Cities)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
